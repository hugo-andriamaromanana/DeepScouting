#### Calcule du nombre de but par équipe

```python


def diviser_par_ponctuation(dataframe):
    # Créer les colonnes "hôte" et "visiteur"
    dataframe['Domicile'] = ''
    dataframe['Exterieur'] = ''
    
    for i, row in dataframe.iterrows():
        chaine = row['game']
        
        # Supprimer les chiffres de la chaîne
        chaine = re.sub('[0-9]', '', chaine)
        
        # Rechercher les ponctuations dans la chaîne
        while chaine.startswith('--') or chaine.startswith('-'):
            chaine = chaine[1:]
        ponctuations = re.findall(r'[^\w\s]', chaine)
        
        if ponctuations:
            # Utiliser la première ponctuation trouvée pour diviser la chaîne en deux groupes
            ponctuation = ponctuations[0]
            
            # Diviser la chaîne en deux groupes à partir de la ponctuation
            groupe1, groupe2 = re.split(re.escape(ponctuation), chaine)
            
            # Retirer les espaces en trop
            groupe1 = groupe1.rstrip()
            groupe2 = groupe2.lstrip()
            
            # Ajouter les données dans les colonnes correspondantes
            dataframe.at[i, 'Domicile'] = groupe1
            dataframe.at[i, 'Exterieur'] = groupe2
    
    return dataframe

# Exemple d'utilisation avec une DataFrame
```

```python
def compare_games_1(dataset):
    unique_games = dataset['game'].unique()
    resultat_match = []
    
    for i in unique_games:
        matching_rows = dataset[dataset['game'] == i]
        
        score = {}
        score['game']=i
       
        if len(matching_rows) > 0:
            count_game1 = matching_rows.loc[:, 'Count'].values[0]
            
            if len(matching_rows) == 1:
                winning_team = matching_rows.loc[matching_rows['game'] == i, 'team'].values[0]
                
                if matching_rows.loc[:, 'Comparison'].values[0] == 'Domicile':
                    result = 'Victoire'
                else:
                    result = 'Défaite'
                
                score['result_match'] = result
                score['domicile_goal'] = count_game1
                score['exterior_goal'] = 0
                
            else:
                game1 = i
                game2 = i
                
                count_game2 = matching_rows.loc[matching_rows['game'] == game2, 'Count'].values[0]
                
                if count_game1 > count_game2:
                    winning_team = matching_rows.loc[matching_rows['game'] == game1, 'team'].values[0]
                    
                    if matching_rows.loc[matching_rows['game'] == game1, 'Comparison'].values[0] == 'Domicile':
                        result = 'Victoire'
                    else:
                        result = 'Défaite'
                    
                    score['result_match'] = result
                    score['domicile_goal'] = count_game1
                    score['exterior_goal'] = count_game2
                
                elif count_game1 < count_game2:
                    winning_team = matching_rows.loc[matching_rows['game'] == game2, 'team'].values[0]
                    
                    if matching_rows.loc[matching_rows['game'] == game2, 'Comparison'].values[0] == 'Domicile':
                        result = 'Victoire'
                    else:
                        result = 'Défaite'
                    
                    score['result_match'] = result
                    score['domicile_goal'] = count_game2
                    score['exterior_goal'] = count_game1
                
                else:
                    result = 'Match null'
                    
                    score['result_match'] = result
                    score['domicile_goal'] = count_game1
                    score['exterior_goal'] = count_game2
            
            resultat_match.append(score)
    
    return resultat_match
```

```python
GER_Bundesliga_22_23_nett= diviser_par_ponctuation(GER_Bundesliga_22_23)
GER_Bundesliga_22_23_match = GER_Bundesliga_22_23_nett[(GER_Bundesliga_22_23_nett['outcome_type'] == 'Successful') & (GER_Bundesliga_22_23_nett['type'] == 'Goal')]
GER_groups = GER_Bundesliga_22_23_match.groupby(['game','team','Domicile','Exterieur']).size().reset_index(name="Count")
GER_groups['Comparison'] = GER_groups.apply(lambda row: 'Exterieur' if row['team'] == row['Exterieur'] else 'Domicile', axis=1)
GER_groups_comp =compare_games_1(GER_groups)

GER_Bundesliga_22_23_match_unique = GER_Bundesliga_22_23_match.drop_duplicates(subset=['game'])
GER_Liga_reorganized = GER_Bundesliga_22_23_match_unique.reset_index(drop=True)

concatenated_GER_Bundesliga = pd.concat([GER_Liga_reorganized,pd.DataFrame(GER_groups_comp)], axis=1)
```
