```python
missing_data_ENG = df_ENG.isnull().sum(axis=0).reset_index()
missing_data_ENG.columns = ['column_name', 'missing_count']
missing_data_ENG['filling_factor'] = (df_ENG.shape[0]
                                - missing_data_ENG['missing_count']) / df_ENG.shape[0] * 100
missing_data_ENG.sort_values('filling_factor').reset_index(drop = True)
```

```python
passes_by_team = df_ITA[(df_ITA['type'] == 'Pass')]['team'].value_counts()
succes = df_ITA[df_ITA["outcome_type"]=="Successful"]
passes_succes = succes[succes['type'] == 'Pass']['team'].value_counts()
per_passes = passes_succes/passes_by_team*100
top_passing_teams = per_passes.head(10)

tackles_by_team = df_ITA[df_ITA['type'] == 'Tackle']['team'].value_counts()
tackles_succes = succes[succes["type"]=="Tackle"]['team'].value_counts()
per_tackles = tackles_succes/tackles_by_team*100
top_tackling_teams = per_tackles.head(10)


aerials_by_team = df_ITA[df_ITA['type'] == 'Aerial']['team'].value_counts()
aerials_succes = succes[succes["type"]=="Aerial"]['team'].value_counts()
per_aerials = aerials_succes/aerials_by_team*100
aerials_teams = per_aerials.head(10)


Goal_by_team = df_ITA[df_ITA['type'] == 'Goal']['team'].value_counts()
shots_by_team = df_ITA[df_ITA['type'] == 'MissedShots']['team'].value_counts() + Goal_by_team
per_goals = Goal_by_team/shots_by_team*100
top_Goal_teams = per_goals.head(10)

balltouch_by_team = df_ITA[df_ITA['type'] == 'BallTouch']['team'].value_counts()
balltouch_succes = succes[succes["type"]=="BallTouch"]['team'].value_counts()
per_balltouch = balltouch_succes/balltouch_by_team*100
per_balltouch_teams = per_balltouch.head(10)

take_on_by_team = df_ITA[df_ITA['type'] == 'TakeOn']['team'].value_counts()
take_on_succes = succes[succes["type"]=="TakeOn"]['team'].value_counts()
per_take_on = take_on_succes/take_on_by_team*100
per_take_on_teams = per_take_on.head(10)

teams=sorted(df_ITA["team"].unique())

l_per_passes = []
l_per_tackles = []
l_per_aerials = []
l_per_goals = []
l_per_balltouch = []
l_per_take_on=[]
for i in range(len(teams)):
    l_per_passes.append(per_passes[i])
    l_per_tackles.append(per_tackles[i])
    l_per_aerials.append(per_aerials[i])
    l_per_goals.append(per_goals[i])
    l_per_balltouch.append(per_balltouch[i])
    l_per_take_on.append(per_take_on[i])




df = pd.DataFrame({
'Group':teams,
'Passes': l_per_passes,
'Tacles': l_per_tackles,
'Aerials': l_per_aerials,
'But/tir': l_per_goals,
'TouchBall' : l_per_balltouch,
'TakeOn':l_per_take_on
})




categories=list(df)[1:]
N = len(categories)

 
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
ax = plt.subplot(111, polar=True)
 
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
plt.xticks(angles[:-1], categories)
 
ax.set_rlabel_position(0)
plt.yticks([30,50,100], ["30","50","100"], color="grey", size=7)
plt.ylim(0,100)

for i in range(len(teams)):
    r = np.round(np.random.rand(),1)
    g = np.round(np.random.rand(),1)
    b = np.round(np.random.rand(),1)
    values=df.loc[i].drop('Group').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=teams[i])
    ax.fill(angles, values, color=[r,g,b], alpha=0.1)
 
 
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.show()
```
